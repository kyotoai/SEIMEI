```python
# Get sample from exp4 and make exp5

import json
import shutil

with open("../exp4/dataset.json") as f:
    dataset = json.load(f)

new_dataset = []
for dict in dataset[:5]:
    shutil.copy("../" + dict["PythonPath"], "../exp5" + dict["PythonPath"][4:])
    shutil.copy("../" + dict["CSVPath"], "../exp5" + dict["CSVPath"][4:])

    dict["PythonPath"] = "exp5" + dict["PythonPath"][4:]
    dict["CSVPath"] = "exp5" + dict["CSVPath"][4:]
    new_dataset.append(dict)

with open("../exp5/dataset.json", "w") as f:
    json.dump(new_dataset, f)
```
